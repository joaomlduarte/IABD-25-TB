```python
# importanto as bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/danielvieira95/Inteligencia_Artificial_2025/refs/heads/main/bases_de_dados/monitoramento_tempo.csv')
```

```python
plt.plot(df['data'], df['humidade'])
```

```python
import datetime
df['data'] = pd.to_datetime(df['data'])
```

```python
plt.plot(df['data'], df['humidade'])
```

```python
plt.plot(df['data'], df['velocidade do vento'])
```

```python
# rezlizando anotações no gráfico
fig = plt.figure(figsize=(15,5))
eixo = fig.add_axes([0,0,1,1])
eixo.plot(df['data'], df['velocidade do vento'], color ='b')
eixo.grid(True)
eixo.set_title('Velocidade do Vento x Ano', fontsize=25, pad=20)
eixo.set_xlabel('Ano', fontsize=20)
eixo.set_ylabel('Velocidade do vento', fontsize=20)
eixo.legend(['data'], loc='best')

# fazendo marcações no grafico
eixo.axhline(max(df['velocidade do vento']), color='k', linestyle='--')
eixo.axhline(min(df['velocidade do vento']), color='k', linestyle='--')

# armazenando valores em x1
x1 = df['data'][df['velocidade do vento'].idxmax()] # armazena a posição em x1
y1 = max(df['velocidade do vento'])
x2 = x1=df['data'][df['velocidade do vento'].idxmax()-7000]
y2 = max(df['velocidade do vento'])-5

# função para fazer anotação no gráfico
eixo.annotate('Máximo', xy=(x1,y1), fontsize=20,xytext=(x2, y2), arrowprops=dict(facecolor='k'))
```

```python
# rezlizando anotações no gráfico
fig = plt.figure(figsize=(15,5))
eixo = fig.add_axes([0,0,1,1])
eixo.plot(df['data'], df['humidade'], color ='b')
eixo.grid(True)
eixo.set_title('Umidade x Ano', fontsize=25, pad=20)
eixo.set_xlabel('Ano', fontsize=20)
eixo.set_ylabel('Umidade', fontsize=20)
eixo.legend(['data'], loc='best')

# fazendo marcações no grafico
eixo.axhline(max(df['humidade']), color='k', linestyle='--')
eixo.axhline(min(df['humidade']), color='k', linestyle='--')

# armazenando valores em x1
x1 = df['data'][df['humidade'].idxmax()] # armazena a posição em x1
y1 = max(df['humidade'])
x2 = x1=df['data'][df['humidade'].idxmax()]
y2 = max(df['humidade'])-5

# função para fazer anotação no gráfico
eixo.annotate('Máximo', xy=(x1,y1), fontsize=20,xytext=(x2, y2), arrowprops=dict(facecolor='k'))
```
