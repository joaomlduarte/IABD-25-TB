```python
#importando as bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Realizar a leitura do meu arquivo CSV
df = pd.read_csv('basededados/monitoramento_tempo.csv', sep=',')
```

```python
# exibindo as primeiras linhas do dataframe
df.head() # exibe as 4 primeiras linhas do dataframe
```

```python
df.info()
```

```python
# importa a biblioteca datetime
import datetime # biblioteca que permite a transformação da data que era um objetct em um datetime
df['data'] = pd.to_datetime(df['data']) # converti a data de object para um datetime
plt.figure(figsize=(15, 8))
plt.plot(df['data'],df['temperatura'])
plt.xlabel('Data e Hora')
plt.ylabel('Temperatura')
```

```python
fig = plt.figure(figsize=(10,8)) #crio uma variável fig para armazenar o tamanho da figura
eixo = fig.add_axes([0,0,1,1]) # define o tamanho dos eixos do grafico
eixo.plot(df['data'], df['temperatura']/10, color='red', lw=0.2, linestyle='-')
eixo.set_title('Temperatura no momento', fontsize=25) # titulo do grafico
eixo.legend(['temperatura'], loc='lower right', fontsize=15)
eixo.set_xlabel('Ano', fontsize=20)
eixo.set_ylabel('Temperatura', fontsize=20)
```

```python
# selecionando o periodo da base de dados
fig = plt.figure(figsize=(15,8))
eixo=fig.add_axes([0,0,1,1])
plt.rcParams['font.size'] = '20'
eixo.plot(df['data'], df['temperatura'], color ='b', lw=0.5, linestyle='-')
eixo.set_xlim(datetime.datetime(2015, 1, 1), datetime.datetime(2015,12,31)) #xlim define o valor incicial e o valor final
eixo.set_title('Temperatura - Ano 2015', fontsize=25)
eixo.legend(['temperatura'], fontsize=20)
eixo.set_xlabel('Ano', fontsize=20)
plt.xticks(rotation=45, fontsize=15, ha='right')
plt.show

```
