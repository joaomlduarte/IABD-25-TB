```python
import matplotlib.pyplot as plt # importa a biblioteca matplotlib
import numpy as np # importa a biblioteca numpy 
```

```python
# Gerando os dados para plotar o grafico
Dados_x = np.linspace(0,10,10) # gera 10 valores de 10 em 10
Dados_y = Dados_x * Dados_x
```

```python
plt.figure(figsize=(10,8), facecolor='gray') # plt. figure cria a figura. figsize tamanho da figura
Ax = plt.axes() # Cria os eixos
Ax.set_facecolor('white') # cor do fundo do grafico
plt.plot(Dados_x, Dados_y, color='red', marker='o', linestyle='dashed', markeredgecolor='black')
# plota o gráfico em x e em y
plt.xlabel('Eixo x', fontsize=16)
plt.ylabel('Eixo y', fontsize=16)
plt.title('Gráfico simples com Matplotlib',fontsize=16)
```
